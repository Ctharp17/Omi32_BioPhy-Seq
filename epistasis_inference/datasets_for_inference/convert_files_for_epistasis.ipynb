Use this script to assign pos1..pos13 and num_mutations for epistasis inference with the BioPhy-Seq datasets.  Run this script once per dataset, changing the in_csv and out_csv variables as needed.  The output files will be used as input for the linear interaction models and global epistasis inference.

```python
import pandas as pd
ba1_in = '../../BioPhy-Seq_measurements/high-throughput_measurements/results_Omi-32/cleaned_datasets/affinity/ba1_affinity_cleaned.csv'
ba1_out = 'ba1_epistasis.csv'
ba4_in = '../../BioPhy-Seq_measurements/high-throughput_measurements/results_Omi-32/cleaned_datasets/affinity/ba4_affinity_cleaned.csv'
ba4_out = 'ba4_epistasis.csv'
wuhan_in = '../../BioPhy-Seq_measurements/high-throughput_measurements/results_Omi-32/cleaned_datasets/affinity/wuhan_affinity_cleaned.csv'
wuhan_out = 'wuhan_epistasis.csv'
expression_in = '../../BioPhy-Seq_measurements/high-throughput_measurements/results_Omi-32/cleaned_datasets/expression/filtered_expression_replicates.csv'
expression_out = 'expression_epistasis.csv'
psr_in = '../../BioPhy-Seq_measurements/high-throughput_measurements/results_Omi-32/cleaned_datasets/polyspecificity/psr_cleaned.csv'
psr_out = 'psr_epistasis.csv'

#select input file to convert for epistasis inference and requisite output file
in_csv  = psr_in
out_csv = psr_out

def find_pos_n(variant, n):
    """Return 1 if the nth (1-based) position in geno == '1', else 0."""
    try:
        return 1 if variant[n-1] == "1" else 0
    except Exception:
        return 0

def sum_digits(digit):
    """Count number of 1’s in geno string."""
    return sum(int(x) for x in str(digit) if x.isdigit())

df = pd.read_csv(in_csv, dtype={"geno": str}).reset_index(drop=True)
df["geno"] = df["geno"].str.zfill(13)

for i in range(1, 14):
    df[f"pos{i}"] = df["geno"].apply(lambda g: find_pos_n(g, i))

df["num_mutations"] = df["geno"].apply(sum_digits)

df_delta = df
print(len(df_delta)/8192)
df_delta.to_csv(out_csv, index=False)
print(f"Added pos1..pos13 + num_mutations.  Wrote → {out_csv}")
```
